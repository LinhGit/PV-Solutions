```python
def create_default_dataset(dataset_path, dataset_type):
    return DataSet(dataset_path)
```

```python
work_dir = "/storage-2/pvextractor-georeferencing/20210510_Schmalenbach/workdir"
cluster_idx = 0 #cluster["cluster_idx"]
dataset_path = os.path.join(work_dir, "mapping", "cluster_{:06d}".format(cluster_idx))

dataset = create_default_dataset(dataset_path, "opensfm")
```

```python
log.setup()
extract_metadata.run_dataset(dataset)
detect_features.run_dataset(dataset)
match_features.run_dataset(dataset)
create_tracks.run_dataset(dataset)
reconstruct.run_dataset(dataset)
```
